```python
import random

play= input("Do you want to play a game (y/n?)")

while play == "y":
  player= input("Choose b/w rock, paper, scissors")
  computer= random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player}")
  print(f"Computer: {computer}")

  if player == computer:
    print("Draw")
  elif player == "rock":
    if computer == "scissors":
      print("Player wins!")
  elif player == "paper":
    if computer == "rock":
      print("Player wins!")
  elif player == "scissors":
    if computer == "paper":
      print("Player wins!")

  else:
    print ("Computer wins!")

  play= input("Do you want to play a game (y/n?)")
```
